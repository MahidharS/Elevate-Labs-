```python
# Handling Missing Values

Some columns had missing data, which can affect analysis and models.

Numeric columns were filled using mean/median to keep the data balanced.

Categorical columns were filled using the most frequent value (mode).

Columns with too many missing values and low importance were dropped.


#Removing Duplicates

Duplicate rows can bias results and cause incorrect insights.
We used .drop_duplicates() to remove repeated records and verified the row count before and after cleaning.


#Converting Data Types

Some columns were stored as text instead of numbers or dates.
We converted them to the correct data types (e.g., numeric, datetime) to allow proper calculations and analysis.


#Creating New Columns (Feature Engineering)

New columns were created using logical rules to improve analysis, such as:

Age groups (Child, Adult, Senior)

Price/Fare bands

Profit margin (if applicable)

This helps models understand patterns better.


#Saving the Cleaned Dataset

After cleaning, the final dataset was saved as a CSV file using .to_csv() so it can be reused for analysis and modeling.
```
